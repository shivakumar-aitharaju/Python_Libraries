#### Plot ly is an interactive visualization library and  Cufflinks connects plotly with pandas


```python
import pandas as pd
import numpy as np
from plotly import __version__
import cufflinks as cf
print(__version__)
```

```python
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
#Data 
df = pd.DataFrame(np.random.randn(100,4), columns=('A B C D').split())
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'], 'Values':[32,43,55]})
df2
```

```python
df.iplot()
```

```python
df.iplot(kind='scatter',x='A',y='B', mode ='markers', size=10)
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.count().iplot(kind='bar')
```

```python
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5], 'y': [10,20,10,40,10], 'z':[200,300,200,500,100 ]})
df3
```

```python
df3.iplot(kind='surface')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5], 'y': [10,20,10,40,10], 'z':[2,3,2,5,1]})
df3.iplot(kind='surface', colorscale='rdylbu')
```

```python
df['A'].iplot(kind='hist', bins=50)
```

```python
df.iplot(kind='hist')
```
