```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x= 'sex', y = 'total_bill',data = tips, hue = 'sex', estimator=np.std)
```

```python
sns.countplot(x = 'sex', data = tips, hue = 'sex')
```

```python
sns.boxplot(x='day', y = 'total_bill', data = tips, hue ='smoker')
```

```python
sns.violinplot(x='day', y='total_bill',data = tips, hue = 'smoker',split=True)
```

```python
sns.stripplot(x='day', y = 'total_bill', data = tips, jitter=True, hue ='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data = tips, hue='day')
```

```python
#Do not use or combine this plots together as you won't understand it.
sns.swarmplot(x='day',y='total_bill',data = tips, hue='day')
sns.violinplot(x='day', y='total_bill',data = tips, hue = 'smoker',split=True, color='violet')
```
