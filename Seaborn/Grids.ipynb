```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
#PairGrid gives more customizations
mygrid = sns.PairGrid(iris)
#mygrid.map(plt.scatter)
mygrid.map_diag(sns.histplot,kde = True)
mygrid.map_lower(sns.kdeplot)
mygrid.map_upper(sns.scatterplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data = tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
# g.map(sns.histplot,'total_bill', kde=True)
```
