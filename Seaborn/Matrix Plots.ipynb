```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')

tips.head()
```

```python
flights.head()
```

```python
#I tried to create heatmap with tips dataset but it was giving error of not converting string to float.
#I was using corr() tips.corr()

#Here is head heatmap with using pivot table of flights dataset
tc = flights.pivot_table(index='month',columns='year',values='passengers')
tc
```

```python
sns.heatmap(tc,cmap='magma',linecolor='white',linewidths=0.1)
```

```python
sns.clustermap(tc,cmap ='coolwarm',linecolor ='black',linewidth=2)

#clustering = similar groups that are close to each other
```
