```python
import matplotlib.pyplot as plt
import pandas as pd
#Here you can apply a new style to change the look of your plots
from qbstyles import mpl_style
df = pd.read_csv("Ice Cream Ratings.csv")
a = df['Date']
b = df['Flavor Rating']
c = df['Overall Rating']
d = df['Texture Rating']
```

```python

# To use this style disable the comments
def plot(dark):
    mpl_style(dark)

plot(dark = True)
```

```python
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
#Functional
plt.plot(x,y)
plt.xlabel("X hai bhau")
plt.ylabel("Y hai bhau")
plt.title("Asach banavla re!")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X label bro")
axes.set_ylabel("Y label Bro")
axes.set_title("Title Bro")
```

```python

fig =  plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.5])
axes2 = fig.add_axes([0.18,0.75,0.2,0.1])

axes1.plot(a,b)
axes2.plot(x,y)
axes1.set_xlabel("X label Bro")
axes1.set_ylabel("Y label Bro")
axes1.set_title("Title BRO")
```

```python
#Sub Plots 
fig, axes = plt.subplots(1,2,figsize=(10,6))

#Axes based on index number
axes[0].plot(x,y)
axes[1].plot(a,b,'g-')   


plt.tight_layout()
```

```python
# Fig size and DPI


fig = plt.figure(figsize=(5,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(a,c)
```

```python
fig,axes = plt.subplots(2,1, figsize=(8,4))

axes[0].plot(a,c,'o-')
axes[0].set_title("Plot 1")
axes[1].plot(a,d,'b-')
axes[1].set_title("Plot 2")
plt.tight_layout()
```
