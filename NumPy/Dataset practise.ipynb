```python
mera_data = pd.read_csv(url)
mera_data
```

```python
# Filtering only setosa from our dataset

setosa_ka_data = mera_data[mera_data['species'] == 'setosa']
setosa_ka_data.head(10)
```

```python
#index variable

x = setosa_ka_data.index

#Filtering 
sepal_length_setosa = setosa_ka_data['sepal_length']

plt.plot(x, sepal_length_setosa, color = '#f77f00', linestyle = ':', linewidth = 3)


# linestyle	            description
# '--' or 'dashed'	    dashed line
# '-.' or 'dashdot'	    dash-dotted line
# ':' or 'dotted'	    dotted line


plt.xlabel('Data Points')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Length of Setosa Iris')
```

```python
# Now take versicolor data 
versicolor_ka_data = mera_data[mera_data['species'] == 'versicolor']
versicolor_ka_data.head()
```

```python
#Filtering 
sepal_length_versicolor = versicolor_ka_data['sepal_length']

sepal_length_versicolor.head(5)

# Now take width of both Setosa and Versicolor
setosa_width = setosa_ka_data['sepal_width']
versicolor_width = versicolor_ka_data['sepal_width']


#Create line of sepal length
plt.plot(x, sepal_length_setosa, label = 'Sepal Length Setosa',
          color = '#03071E', linestyle = ':' , linewidth = 2)
plt.plot(x, sepal_length_versicolor, label = 'Sepal Length Versicolor', 
         color = '#DC2F02', linestyle = '-' , linewidth = 2)


#Create line of Sepal Width 
plt.plot(x, setosa_width, label = 'Sepal Width Setosa', 
         color = '#6A040F', linestyle = '--' , linewidth = 2 )
plt.plot(x, versicolor_width, label = 'Sepal Width Versicolor',
          color = '#FFBA08', linestyle = ':' , linewidth = 2)

# Plotting a figure of width 6 and height 3

plt.legend()
plt.show()

```

```python
#Bar Chart
species = mera_data['species'].unique()

sepal_length_means = [mera_data[mera_data['species'] == spec]['sepal_length'].mean() for spec in species]



# Create a bar plot
plt.bar(species, sepal_length_means, color=['#8d99ae', '#ef233c', '#735751'])

# Add labels and a title
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.title('Average Sepal Length by Species')

# Display the plot
plt.show()
```
